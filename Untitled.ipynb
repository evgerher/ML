```python
import pandas as pd
import numpy as np
import sklearn as sk
from matplotlib import pyplot as plt
```

```python
iris = pd.read_csv("datasets/iris.csv")
nparray = iris.values
features = nparray[:,:4]
target = nparray[:,4:5]
```

```python
def splitter(features, targets, test_fraction=0.7, shift=0):
    if shift > 100:
        return None
    shift /= 100 # percents
    X, Y = sk.utils.shuffle(features, target, random_state=0)
    length = len(X)
    start = int(length * shift)
    end = int(start + length * test_fraction)
    shuffled_x = np.array([])
    shuffled_y = np.array([])
    test_x = np.array([])
    test_y = np.array([])
    
    
    shuffled_x = np.array([X[int(_) % length] for _ in range(start, end)])
    shuffled_y = np.array([Y[int(_) % length] for _ in range(start, end)])
    
    if end < start:
        test_x = np.array([X[_] for _ in range(end, start)])
        test_y = np.array([Y[_] for _ in range(end, start)])
    else:
        t1x = np.array([X[_] for _ in range(0, start)])
        t2x = np.array([X[_] for _ in range(end, length)])
        t1y = np.array([Y[_] for _ in range(0, start)])
        t2y = np.array([Y[_] for _ in range(end, length)])
#         print(t1x)
#         print("#"*10)
#         print(t2x)
        test_x = np.concatenate((t1x, t2x), axis=0)
        test_y = np.concatenate((t1y, t2y), axis=0)

    return shuffled_x, shuffled_y, test_x, test_y
```

```python
X, Y, Tx, Ty = splitter(features, target, 0.7, 10)
print(Ty)
```
